# Algorithme Génétique

```python
from PolynomProblem import PolynomProblem
from PolynomInvidual import PolynomInvidual
from AlgoGen import AlgoGen
import matplotlib.pyplot as plt
from random import randint

class Devine:
    def __init__(self,nb_const,max_value):
        self.l = [randint(0,max_value) for _ in range(nb_const)]
        
    def get(self,n):
        return self.l[n]
    
    def f(self,i):
        res = 0
        for j in range(len(self.l)):
            res += self.l[j]*(i**j)
        return res

N = 10
MAX = 10
NB_CST = 5

mystere = Devine(NB_CST,MAX)
expected = [mystere.f(i) for i in range(N)]

pb = PolynomProblem(mystere.f,NB_CST,MAX,N) #funct,nb_degree,max_value,n
algo = AlgoGen(pb,50,0,0) #problem,population_size,crossover_rate,mutation_probability
algo.run(1000)
```
